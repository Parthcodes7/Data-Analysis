```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
plt.plot(x, y)
plt.title("Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-label")
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(y,x)
plt.subplot(2,2,3)
plt.plot(x**2,x**2)
plt.subplot(2,2,4)
plt.plot(x**3,y**2)
```

```python
fig = plt.figure(figsize=(6,8),dpi = 100)
axis1 = fig.add_axes([0.1,0.1,0.8,0.8])
axis1.plot(x,y)
axis2 = fig.add_axes([0.2,0.5,.3,0.3])
axis2.plot(x,y)
```
