```python
import numpy as np
```

```python
arr = np.array([1,2,3,4,5])
x  = arr.copy()
arr[0] = 42

print(arr)
print(x)

```

```python
arr = np.array([1,2,3,4,5])
x  = arr.view()
arr[0] = 42

print(arr)
print(x)
```

```python
arr1 = np.array([1,2,3])

for x in arr1:
    print(x)
```

```python
arr2 = np.array([1,2,3,4])
arr3 = np.array([5,6,7,8])

arr4 = np.stack((arr2, arr3), axis = 1)

print(arr4)
```

```python
arr2 = np.array([1,2,3,4])
arr3 = np.array([5,6,7,8])

arr4 = np.hstack((arr2, arr3))

print(arr4)
```

```python
arr2 = np.array([1,2,3,4])
arr3 = np.array([5,6,7,8])

arr4 = np.dstack((arr2, arr3))

print(arr4)
```

```python
arr5 = np.array([1,2,3,4,5,6])
new_arr = np.array_split(arr5, 3)


print(new_arr[0])
print(new_arr[1])
print(new_arr[2])
```

```python
arr6 = np.array([1,2,3,4,5,4,4])
x = np.where(arr6 == 4)

print(x)
```

```python
arr6 = np.array([1,2,3,4,5,4,4])
x = np.where(arr6%2 == 0)

print(x)
```

```python
arr7 = np.array([6,7,8,9])
x = np.searchsorted(arr7,7)

print(x)
```

```python
arr7 = np.array([6,7,8,9])
x = np.searchsorted(arr7,7, side = 'right')

print(x)
```

```python
arr8 = np.array([1,3,5,7])

x1 = np.searchsorted(arr8, [2,4,6])

print(x1)
```
